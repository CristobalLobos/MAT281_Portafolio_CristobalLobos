EL conjunto de datos se denomina `company_sales_data.csv`, el cual contiene información tal como: número del mes, unidades, precio, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv(os.path.join("data","company_sales_data.csv"))
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
total_profit = df['total_profit'] #leer datos
months = df['month_number']

fig = plt.figure(figsize=(14, 9)) #creacion de graficos
ax1 = fig.add_subplot(2, 1, 1) 
ax2 = fig.add_subplot(2, 1, 2, sharex=ax1)


ax1.plot(months, total_profit, color = 'black', linewidth = 1)#grafico lineal
ax1.set_xlabel('Mes (month_number)', fontsize=16)
ax1.set_ylabel('Ganancia total (total_profit)', fontsize=16)

ax2.scatter(months,total_profit, color = 'blue', s=8) #grafico dispersion
ax2.set_xlabel('Mes (month_number)', fontsize=16)
ax2.set_ylabel('Ganancia total (total_profit)', fontsize=16)


plt.show()
```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
df1= df.drop(['total_units', 'total_profit'], axis = 1)

plt.figure(figsize=(16, 10)) #creacion de grafico

for y in list(df1.columns[1:]): #se omite la columna de meses (corresponde a la primera columna del dataframe)
    sns.lineplot(x = 'month_number', y = y, linewidth = 2, data = df1, label=str(y)) #se iterando graficando cada columna produciendo un grafico multilinea
    
plt.xlabel("Mes (en numero)", fontsize = 20) #etiquetas 
plt.ylabel("Ventas totales (por objeto)", fontsize = 20)
plt.show()
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
ax = df[['facecream','facewash']].plot(kind='bar', figsize=(15, 10), legend=True, fontsize=15) #creacion del grafico utilizando los datos pedidos
ax.set_xlabel("Mes (en numero)", fontsize=20) #etiquetas de los ejes
ax.set_ylabel("Ventas", fontsize=20)
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
df_ventas = df.drop(['total_units', 'total_profit', 'month_number'], axis = 1) #ventas de productos
plt.figure(figsize=(15, 9)) #creacion grafico
sns.boxplot(data=df_ventas) #grafico de caja

plt.xlabel("Producto", fontsize = 20) #etiquetas de los ejes
plt.ylabel("Numero de ventas", fontsize = 20)
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
df_ventas = df.drop(['total_units', 'total_profit', 'month_number'], axis = 1) #lectura de los datos pedidos
textprops = {"fontsize":20} #tamaño de la fuente
plt.figure(figsize=(17, 10)) #creacion del grafico
df1 = df_ventas.sum() #suma de los valores para obtener las ventas totales de cada producto
plt.pie(x=df1, labels=df1.index,  autopct='%.2f%%', textprops = textprops) #autopct presenta los valores en porcentaje
```
